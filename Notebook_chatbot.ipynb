```python
# Chargement des documents PDF

from pathlib import Path
from langchain.document_loaders import PyMuPDFLoader

pdf_folder = Path("./data")
pdf_files = list(pdf_folder.glob("*.pdf"))

raw_documents = []
for pdf in pdf_files:
    loader = PyMuPDFLoader(str(pdf))
    docs = loader.load()
    raw_documents.extend(docs)
```

```python
print(f" Documents chargés : {len(raw_documents)}")
```

```python
# Découpage des documents
from langchain.text_splitter import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
documents = splitter.split_documents(raw_documents)
```

```python
len(documents)
```

```python
#Embeddings & Vectorisation avec OpenAI + Chroma
import os
from dotenv import load_dotenv
from langchain_openai import OpenAIEmbeddings
from langchain.vectorstores import Chroma

load_dotenv()
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")

# Test avec 50 chunks pour voir si le kernel tient
documents_subset = documents[:30]

# embedding_model = OpenAIEmbeddings()
# vectorstore = Chroma.from_documents(documents_subset, embedding_model, persist_directory="chroma_fiscalite")
# vectorstore.persist()

embedding_model = OpenAIEmbeddings()
vectorstore = Chroma.from_documents(documents, embedding_model, persist_directory="chroma_fiscalite")
vectorstore.persist()

print("Index vectoriel Chroma créé et sauvegardé.")
```
